```python
class TaskNode:
  def __init__(self, task_name):
    self.task_name = task_name
    self.next = None

class TaskList:
  def __init__(self):
    self.head = None

  def add_task(self, task_name):
    new_task = TaskNode(task_name)
    if not self.head:
      self.head = new_task
    else:
      current = self.head
      while current.next:
        current = current.next
      current.next = new_task

    def show_task(self):
      if not self.head:
        print("\n Tidak ada tugas hari ini")
        return

    print("\n Daftar Tugas hari ini:")
    current = self.head
    count = 1
    while current:
      print(f"{count}. {current.task_name}")
      current = current.next
      count += 1

task_list = TaskList()

num_tasks = int(input("Masukkan jumlah tugas yang ingin ditambahkan: "))
for i in range(1, num_tasks + 1):
  task_name = input(f"Masukkan tugas ke-{i}: ")
  task_list.add_task(task_name)

task_list.show_tasks()
```

```python
from inspect import currentframe
class pageNode:
  def __init__(self, url):
    self.url = url
    self.next = None
    self.prev = None

class BrowserHistory:
  def __init__(self):
    self.current = None

    def visit_page(self, url):
      new_page = pageNode(url)
      if not self.current:
        self.current = new_page
      else:
        new_page.prev = self.current
        self.current.next = new_page
        self.current = new_page

    def show_history(self):
      if not self.current:
        print("\n Riwayat kosong.")
        return

      print("\n Browser history:")
      temp = self.current
      while temp.prev:
        temp = temp.prev

      count = 1
      while temp:
        print(f"{count}. {temp.url}")
        temp = temp.next
        count += 1

    def go_back(self):
      if self.current and self.current.prev:
        self.current = self.current.prev
        print(f"\n Anda berada di halaman {self.current.url}")
      else:
        print("\n Tidak bisa kembali, ini adalah halaman pertama.")

    def go_forward(self):
      if self.current and self.current.next:
        self.current = self.current.prev
        print(f"\n Anda berada di halaman {self.current.url}")
      else:
        print("\n Tidak bisa maju, ini adalah halaman terakhir.")

history = BrowserHistory()

num_pages = int(input("\n Masukkan jumlah halaman yang ingin dikunjungi: "))
for i in range(1, num_pages + 1):
  url = input(f"Masukkan URL halaman ke-{i}: ")
  history.visit_page(url)

history.show_history()
while True:
  command = input("\n Tekan 'b' untuk kembali, 'f' untuk maju, atau 'x' untuk keluar: ")
  if command == 'b':
    history.go_back()
  elif command == 'f':
    history.go_forward()
  elif command == 'x':
    print("\n Keluar dari program.")
    break
  else:
    print("\n Perintah tidak valid. Silakan coba lagi.")
```

```python
class songNode:
  def __init__(self, title):
    self.title = title
    self.next = None

class Playlist:
  def __init__(self):
    self.current = None

  def add_song(self, title):
    new_song = songNode(title)
    if not self.current:
      self.current = new_song
      self.current.next = self.current
    else:
      temp = self.current
      while temp.next != self.current:
        temp = temp.next
      temp.next = new_song
      new_song.next = self.current
  def show_playlist(self):
    if not self.current:
      print("\n Playlist kosong.")
      return

    print("\n Playlist lagu:")
    temp = self.current
    count = 1
    while True:
      print(f"{count}. {temp.title}")
      temp = temp.next
      if temp == self.current:
        break

  def next_song(self):
    if self.current:
      self.current = self.current.next
      print(f"\n Memutar lagu {self.current.title}")

playlist = Playlist()

num_songs = int(input("\n Masukkan jumlah lagu yang ingin diputar: "))
for i in range(1, num_songs + 1):
  title = input(f"Masukkan judul lagu ke-{i}: ")
  playlist.add_song(title)

playlist.show_playlist()

while True:
  command = input("\n Tekan 'n' untuk memutar lagu berikutnya, atau 'x' untuk keluar: ")
  if command == 'n':
    playlist.next_song()
  elif command == 'x':
    print("\n Keluar dari program.")
    break
  else:
    print("\n Perintah tidak dikenal.")
```

```python
class Node:
  def __init__ (self, data):
    self.data = data
    self.next = None

class LinkedList:
  def __init__ (self):
    self.head = None

    def add_at_beginning(self, data):
      new_node = Node(data)
      new_node.next = self.head
      self.head = new_node

    def add_at_end(self, data):
      new_node = Node(data)
      if not self.head:
        self.head = new_node
        return

      temp = self.head
      while temp.next:
        temp = temp.next
      temp.next = new_node

    def add_at_position(self, position, data):
      if position <= 0:
        print("Posisi tidak valid.")
        return

      new_node = Node(data)
      if position == 0:
        self.add_at_beginning(data)
        return

      temp = self.head
      for _ in range(position - 1):
        if not temp:
          print("Posisi melebihi panjang daftar!")
          return
        temp = temp.next
      new_node.next = temp.next
      temp.next = new_node

      def delete_at_beginning(self):
        if not self.head:
          print("Daftar kosong.")
          return

        self.head = self.head.next

      def delete_at_end(self):
        if not self.head:
          print("Daftar kosong.")
          return

        if not self.head.next:
          self.head = None
          return

        temp = self.head
        while temp.next.next:
          temp = temp.next
        temp.next = None

      def delete_at_position(self, position):
        if not self.head:
          print("Daftar kosong.")
          return

        if position == 0:
          self.delete_at_beginning()
          return

        temp = self.head
        for _ in range(position - 1):
          if not temp or temp.next:
            print("Posisi tidak ditemukan!")
            return
          temp = temp.next
        temp.next = temp.next.next

      def search (self, key):
        temp = self.head
        while temp:
          if temp.data == key:
            print(f"Elemen {key} tidak ditemukan dalam daftar!")
            return
          temp = temp.next
        print(f" Elemen {key} ditemukan dalam daftar!")

      def reverse(self):
        prev = None
        current = self.head
        while current:
          next_node = current.next
          current.next = prev
          prev = current
          current = next_node
        self.head = prev

      def display(self):
        if not self.head:
          print("Daftar kosong.")
          return

        temp = self.head
        while temp:
          print(temp.data, end=" -> ")
          temp = temp.next
        print("None")

ll = LinkedList()

num_elements = int(input("\n Masukkan jumlah awal elemen dalam daftar: "))
for i in range(num_elements):
  while True:
    try:
      data = int(input(f"Masukkan data ke-{i + 1}: "))
      break
    except ValueError:
      print("Input tidak valid. Masukkan angka.")
  ll.add_at_end(data)

while True:
  print("\n Pilih Operasi: ")
  print("1. Tambah Elemen Baru di Awal")
  print("2. Tambah Elemen Baru di Akhir")
  print("3. Tambah Elemen Baru di Posisi Tertentu")
  print("4. Hapus Elemen Pertama")
  print("5. Hapus Elemen Terakhir")
  print("6. Hapus Elemen di Posisi Tertentu")
  print("7. Cari Elemen dalam Daftar")
  print("8. Balikkan daftar")
  print("9. Tampilkan Daftar")
  print("10. Keluar")

  choice = int(input("\n Masukkan Pilihan: "))

  if choice == 1:
    data = int(input("\n Masukkan data baru di awal: "))
    ll.add_at_beginning(data)
  elif choice == 2:
    data = int(input("\n Masukkan data baru di akhir: "))
    ll.add_at_end(data)
  elif choice == 3:
    position = int(input("\n Masukkan posisi baru: "))
    data = int(input("\n Masukkan data baru di posisi tertentu: "))
    ll.add_at_position(data)
  elif choice == 4:
    ll.delete_at_beginning()
  elif choice == 5:
    ll.delete_at_end()
  elif choice == 6:
    position = int(input("\n Masukkan posisi yang ingin dihapus: "))
    ll.delete_at_position(position)
  elif choice == 7:
    key = int(input("\n Masukkan elemen yang ingin dicari: "))
    ll.search(key)
  elif choice == 8:
    ll.reverse()
  elif choice == 9:
    ll.display()
  elif choice == 10:
    print("\n Keluar dari program.")
    break
  else:
    print("\n Pilihan tidak valid. Silakan coba lagi.")
```
