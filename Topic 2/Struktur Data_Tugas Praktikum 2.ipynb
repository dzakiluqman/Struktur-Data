```python
class Node:
  def __init__(self, name):
    self.name = name
    self.next = None

class CircularLinkedList:
  def __init__(self):
    self.head = None

  def add_node(self, name):
    new_node = Node(name)
    if not self.head:
      self.head = new_node
      self.head.next = self.head
    else:
      temp = self.head
      while temp.next != self.head:
        temp = temp.next
      temp.next = new_node
      new_node.next = self.head

  def schedule(self, days):
    current = self.head
    print("\nJadwal piket satpam selama 2 minggu: \n")
    for day in range(1, days + 1):
      print(f"Hari {day}: {current.name}")
      current = current.next

satpam_list = CircularLinkedList()
for i in range(4):
  name = input(f"Masukkan nama satpam ke-{i+1}: ")
  satpam_list.add_node(name)

satpam_list.schedule(14)
```
