```python
def quick_sort(arr, key):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[len(arr) // 2]
        left = [x for x in arr if x[key] < pivot[key]]
        middle = [x for x in arr if x[key] == pivot[key]]
        right = [x for x in arr if x[key] > pivot[key]]
        return quick_sort(left, key) + middle + quick_sort(right, key)

customers = [
    {"nama": "Bambang", "umur": 25, "total_belanja": 500000},
    {"nama": "Agus", "umur": 22, "total_belanja": 750000},
    {"nama": "Dono", "umur": 30, "total_belanja": 300000},
    {"nama": "Iwan", "umur": 27, "total_belanja": 1000000}
]

print("Pilih metode sorting:")
print("1. Nama (A-Z)")
print("2. Umur")
print("3. Total Belanja")
pilihan = input("Masukkan pilihan (1/2/3): ")

if pilihan == "1":
    sorted_customers = quick_sort(customers, "nama")
elif pilihan == "2":
    sorted_customers = quick_sort(customers, "umur")
elif pilihan == "3":
    sorted_customers = quick_sort(customers, "total_belanja")
else:
    print("Pilihan tidak valid!")
    exit()

print("\nData pelanggan setelah diurutkan:")
for c in sorted_customers:
    print(f"Nama: {c['nama']}, Umur: {c['umur']}, Total Belanja: Rp{c['total_belanja']}")
```
