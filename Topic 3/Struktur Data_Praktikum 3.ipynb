```python
class TextEditor:
  def __init__(self):
    self.history = []

  def write(self, text):
    self.history.append(text)

  def undo(self):
    if self.history:
      return f"Undo: Menghapus '{self.history.pop()}'"
    else:
      return "Tidak ada yang dapat di-undo"

editor = TextEditor()

editor.write("Halo, ini teks pertama.")
editor.write("Ini adalah teks kedua.")
editor.write("Ini adalah teks ketiga.")

print(editor.undo())
print(editor.undo())
```

```python
class UndoCalculator:
  def __init__(self, value):
    self.stack = [value]

  def tambah(self, value):
    self.stack.append(self.stack[-1] + value)

  def kurang(self, value):
    self.stack.append(self.stack[-1] - value)

  def kali(self, value):
    self.stack.append(self.stack[-1] * value)

  def undo(self):
    if self.stack:
      return f"Undo: Menghapus '{self.stack.pop()}'"
    else:
      return "Tidak ada yang dapat di-undo"

value = int(input("Masukkan dua digit terakhir NPM Anda: "))
kalkulator = UndoCalculator(value)

kalkulator.tambah(10)
print("Setelah tambah 10:", kalkulator.stack[-1])

kalkulator.kurang(5)
print("Setelah kurang 5:", kalkulator.stack[-1])

kalkulator.kali(2)
print("Setelah kali 2:", kalkulator.stack[-1])

print(kalkulator.undo())
print(kalkulator.undo())
```

```python
from collections import deque

class BrowserHistory:
  def __init__(self):
    self.history = deque()

  def visit(self, url):
    self.history.append(url)
    print(f"Berhasil mengakses '{url}'")

  def back(self):
    if self.history:
      print(f"Berhasil kembali ke '{self.history.pop()}'")
    else:
      print("Tidak ada yang dapat dikembalikan")

browser = BrowserHistory()
browser.visit("google.com")
browser.visit("youtube.com")
browser.visit("github.com")

browser.back()
browser.back()
```

```python
from collections import deque

class NavigasiHistory:
  def __init__(self):
    self.history = deque()

  def visit(self, fitur):
    self.history.append(fitur)
    print(f"Berhasil mengakses fitur '{fitur}'")

  def back(self):
    if self.history:
      print(f"Berhasil kembali ke fitur '{self.history.pop()}'")
    else:
      print("Tidak ada yang dapat dikembalikan")

mobile = NavigasiHistory()
mobile.visit("M16")
mobile.visit("E16")
mobile.visit("I16")

mobile.back()
mobile.back()
```

```python
from queue import LifoQueue

class BackupSystem:
  def __init__(self):
    self.backup = LifoQueue()

  def backup(self, version):
    self.backup.put(version)

  def latest_backup(self):
    if not self.backup.empty():
      return f"Versi Terbaru: {self.backup.queue[-1]}"
    return "Tidak ada yang dapat di-backup"

backup_system = BackupSystem()
backup_system.backup("Backup_1")
backup_system.backup("Backup_2")
backup_system.backup("Backup_3")

print(backup_system.latest_backup())
```

```python
from queue import LifoQueue

class DraftEmail:
  def __init__(self):
    self.draft_queue = LifoQueue()

  def draft(self, version):
    self.draft_queue.put(version)

  def latest_draft(self):
    if not self.draft_queue.empty():
      return f"Draft Email Terbaru: {self.draft_queue.queue[-1]}"
    return "Tidak ada yang di draft."

email = DraftEmail()
email.draft("D16")
email.draft("Z16")
email.draft("A16")

print(email.latest_draft())
```

```python
class CS:
  def __init__(self):
    self.ticket = []

  def add_ticket(self, ticket):
    self.ticket.append(ticket)

  def process_ticket(self):
    if not self.is_empty():
      return f"Memproses tiket: {self.ticket.pop(0)}"
    return "Tidak ada yang dapat diproses"

  def is_empty(self):
    return len(self.ticket) == 0

service = CS()
service.add_ticket("Tiket Film M16")
service.add_ticket("Tiket Film E16")
service.add_ticket("Tiket Film I16")

print(service.process_ticket())
print(service.is_empty())
```
